```python
%matplotlib inline

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
```

```python
def load_df(fp):
    df = pd.read_csv(fp, names=['Function', 'Container', 'K', 'T', 'time_ns'], header=0)
    df['time_us'] = df.time_ns / 1000
    return df
```

```python
HMMBase_df = load_df('HMMBase/HMMBase_v0.0.10_1552152589.csv')
HMMBase_df.head()
```

```python
hmmlearn_df = load_df('hmmlearn/hmmlearn_v0.2.1_1552152210.csv')
hmmlearn_df.head()
```

```python
def plot_time_vs_k(df):
    plt.figure(figsize=(6,4))
    # TODO: Default figure size
    # TODO: Display n_samples
    sns.lineplot(x='K', y='time_us', hue='Container', style='Function', data=df[df['T'] == 1000])
    plt.xticks(sorted(df.K.unique()));
    plt.ylabel('Time (µs)')
```

```python
def plot_time_vs_T(df):
    plt.figure(figsize=(6,4))
    # TODO: Default figure size
    # TODO: Display n_samples
    sns.lineplot(x='T', y='time_us', hue='Container', style='Function', data=df[df['K'] == 10])
    plt.xscale('log')
    plt.yscale('log')
    plt.xticks(sorted(df['T'].unique()));
    plt.ylabel('Time (µs)')
```

```python
plot_time_vs_k(HMMBase_df)
plot_time_vs_k(hmmlearn_df)
```

```python
plot_time_vs_T(HMMBase_df)
plot_time_vs_T(hmmlearn_df)
```
